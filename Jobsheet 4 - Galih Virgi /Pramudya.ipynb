#Praktikum 1: Inheritance Dasar



```python
# Kelas Induk
class Kendaraan:
    def __init__(self, merk):
        self.merk = merk

    def mulai_mesin(self):
        print(f"Mesin kendaraan {self.merk} dinyalakan.")

# Kelas Anak (mewarisi dari Kendaraan)
class Mobil(Kendaraan):
    def __init__(self, merk, warna):
        # Memanggil init kelas induk (akan dibahas di Praktikum 2)
        # Untuk sekarang, kita set merk secara manual
        self.merk = merk  # Mewarisi 'merk' tapi di-set ulang di sini
        self.warna = warna  # Atribut khusus Mobil

    def info_mobil(self):
        print(f"Ini adalah mobil {self.merk} berwarna {self.warna}.")

# --- Kode Utama ---
if __name__ == "__main__":
    mobil_tesla = Mobil("Tesla Model S", "Merah")

    # Memanggil metode dari kelas anak
    mobil_tesla.info_mobil()

    # Memanggil metode yang diwarisi dari kelas induk
    mobil_tesla.mulai_mesin()

    # Mengakses atribut yang diwarisi (dan di-set di anak)
    print(f"Merk mobil: {mobil_tesla.merk}")
```

#Praktikum 2: Menggunakan super() dalam Konstruktor

```python
# --- Kelas Induk ---
class Person:
    def __init__(self, nama, usia):
        print(f"(Memanggil __init__ Person untuk '{nama}')")
        self.nama = nama
        self.usia = usia

    def perkenalkan_diri(self):
        print(f"Halo, nama saya {self.nama}, usia saya {self.usia} tahun.")

# --- Kelas Anak (mewarisi dari Person) ---
class Student(Person):
    def __init__(self, nama, usia, student_id, jurusan):
        print(f"(Memanggil __init__ Student untuk '{nama}')")
        # Memanggil __init__ dari kelas Person untuk inisialisasi nama dan usia
        super().__init__(nama, usia)

        # Menambahkan atribut khusus Student
        self.student_id = student_id
        self.jurusan = jurusan
        print(f"(Inisialisasi atribut Student selesai untuk '{nama}')")

    def info_akademik(self):
        print(f"ID Mahasiswa: {self.student_id}")
        print(f"Jurusan: {self.jurusan}")

    # Override perkenalkan_diri untuk menambahkan info student
    def perkenalkan_diri(self):
        super().perkenalkan_diri()  # Panggil versi Person
        print(f"Saya adalah mahasiswa dengan ID {self.student_id}, jurusan {self.jurusan}.")

# --- Kode Utama ---
if __name__ == "__main__":
    # Membuat objek Person
    dosen = Person("Pak Anton", 45)
    print("-" * 20)
    dosen.perkenalkan_diri()
    print("\n" + "=" * 30 + "\n")

    # Membuat objek Student
    mahasiswa = Student("Dewi", 20, "MHS001", "Teknik Komputer")
    print("-" * 20)

    # Panggil versi override di Student
    mahasiswa.perkenalkan_diri()
    print("-" * 20)

    # Panggil metode khusus Student
    mahasiswa.info_akademik()

    # Cek atribut yang diwarisi
    print(f"\nUsia mahasiswa {mahasiswa.nama}: {mahasiswa.usia}") # usia diwarisi dari Person
```

#Praktikum 3: Method Overriding dan super()


```python
# --- Kelas Induk ---
class Hewan:
    def __init__(self, nama):
        self.nama = nama

    def bersuara(self):
        print(f"{self.nama} mengeluarkan suara generik.")

    def tampilkan_info(self):
        print(f"Ini adalah hewan bernama {self.nama}.")

# --- Kelas Anak ---
class Kucing(Hewan):
    def __init__(self, nama, ras):
        super().__init__(nama)  # Panggil init induk
        self.ras = ras

    # Method Overriding untuk bersuara
    def bersuara(self):
        print(f"{self.nama} (Kucing) mengeong: Meow!")

    # Method Overriding untuk tampilkan_info, memanggil versi induk
    def tampilkan_info(self):
        super().tampilkan_info()  # Panggil implementasi dari Hewan
        print(f"Ini adalah kucing ras {self.ras}.")  # Tambahkan info spesifik

# --- Kode Utama ---
if __name__ == "__main__":
    hewan_umum = Hewan("Makhluk")
    kucing_persia = Kucing("Puspus", "Persia")

    print("Info Hewan Umum:")
    hewan_umum.tampilkan_info()
    hewan_umum.bersuara()

    print("-" * 20)

    print("Info Kucing Persia:")
    kucing_persia.tampilkan_info()  # Memanggil versi Kucing (yang juga memanggil induk)
    kucing_persia.bersuara()       # Memanggil versi Kucing
```

#Praktikum 4: Multi-level Inheritance (Pewarisan Bertingkat)

```python
# Kelas Induk Level 1
class Organisme:
    def __init__(self, nama):
        self.nama = nama
        print(f"Organisme '{self.nama}' diciptakan.")

    def bernapas(self):
        print(f"{self.nama} sedang bernapas.")

# Kelas Anak Level 1 (mewarisi dari Organisme)
class Hewan(Organisme):
    def __init__(self, nama, jenis_makanan):
        # Panggil init Organisme
        super().__init__(nama)
        self.jenis_makanan = jenis_makanan
        print(f"Hewan '{self.nama}' adalah {self.jenis_makanan}.")

    def bergerak(self):
        print(f"{self.nama} sedang bergerak.")

# Kelas Anak Level 2 (mewarisi dari Hewan)
class Mamalia(Hewan):
    def __init__(self, nama, jenis_makanan, jumlah_kaki):
        # Panggil init Hewan
        super().__init__(nama, jenis_makanan)
        self.jumlah_kaki = jumlah_kaki
        print(f"Mamalia '{self.nama}' memiliki {self.jumlah_kaki} kaki.")

    def menyusui(self):
        print(f"{self.nama} sedang menyusui.")

# Kode Utama
if __name__ == "__main__":
    kucing = Mamalia("Kucing Persia", "Karnivora", 4)
    print("-" * 20)

    # Memanggil metode dari kelas Mamalia
    kucing.menyusui()

    # Memanggil metode yang diwarisi dari kelas Hewan
    kucing.bergerak()
    print(f"Jenis makanan: {kucing.jenis_makanan}")

    # Memanggil metode yang diwarisi dari kelas Organisme
    kucing.bernapas()
    print(f"Nama organisme: {kucing.nama}")
```

```python
# === BAGIAN HACKER: Instalasi Locale Indonesia di Sistem Colab ===
import os
print("Sedang mengonfigurasi sistem... Mohon tunggu sebentar.")
# Mengunduh dan menginstal data bahasa Indonesia ke sistem Linux Colab
!apt-get install -y locales > /dev/null
!locale-gen id_ID.UTF-8 > /dev/null

# Mengatur environment variable agar sistem mengenali bahasa Indonesia
os.environ['LC_ALL'] = 'id_ID.UTF-8'

import locale
try:
    locale.setlocale(locale.LC_ALL, 'id_ID.UTF-8')
    print("Sistem berhasil dikonfigurasi ke id_ID.UTF-8!\n")
except locale.Error:
    print("Gagal memuat locale, menggunakan fallback format.")

# --- Fungsi Helper ---
def format_rupiah(angka):
    """Menggunakan library locale yang sudah diinstal di sistem"""
    try:
        return locale.currency(angka, grouping=True, symbol='Rp ')
    except:
        # Fallback jika locale tetap tidak terbaca
        return f"Rp {angka:,.0f}".replace(',', '.')

# --- Kelas Induk ---
class Pegawai:
    def __init__(self, nama, id_pegawai, gaji_pokok):
        self.nama = nama
        self.id_pegawai = id_pegawai
        self.gaji_pokok = gaji_pokok

    def hitung_gaji(self):
        return self.gaji_pokok

    def tampilkan_info(self):
        print(f"ID Pegawai   : {self.id_pegawai}")
        print(f"Nama         : {self.nama}")
        print(f"Gaji Pokok   : {format_rupiah(self.gaji_pokok)}")

# --- Kelas Anak 1 ---
class Manager(Pegawai):
    def __init__(self, nama, id_pegawai, gaji_pokok, tunjangan_jabatan):
        super().__init__(nama, id_pegawai, gaji_pokok)
        self.tunjangan_jabatan = tunjangan_jabatan

    def hitung_gaji(self):
        return super().hitung_gaji() + self.tunjangan_jabatan

    def tampilkan_info(self):
        print("--- Info Manager ---")
        super().tampilkan_info()
        print(f"Tunjangan    : {format_rupiah(self.tunjangan_jabatan)}")
        print(f"Total Gaji   : {format_rupiah(self.hitung_gaji())}")

# --- Kelas Anak 2 ---
class StafTeknis(Pegawai):
    def __init__(self, nama, id_pegawai, gaji_pokok, keahlian, bonus_keahlian):
        super().__init__(nama, id_pegawai, gaji_pokok)
        self.keahlian = keahlian
        self.bonus_keahlian = bonus_keahlian

    def hitung_gaji(self):
        return super().hitung_gaji() + self.bonus_keahlian

    def tampilkan_info(self):
        print("--- Info Staf Teknis ---")
        super().tampilkan_info()
        print(f"Keahlian     : {self.keahlian}")
        print(f"Bonus Skill  : {format_rupiah(self.bonus_keahlian)}")
        print(f"Total Gaji   : {format_rupiah(self.hitung_gaji())}")

# --- Eksekusi ---
if __name__ == "__main__":
    manager1 = Manager("Budi Santoso", "M001", 10000000, 5000000)
    staf1 = StafTeknis("Citra Lestari", "S001", 7000000, "Python Programming", 1500000)
    pegawai_biasa = Pegawai("Rian", "P005", 5000000)

    print("=== SISTEM PENGGAJIAN (LOCALE MODE) ===\n")

    manager1.tampilkan_info()
    print("-" * 35)

    staf1.tampilkan_info()
    print("-" * 35)

    print("--- Info Pegawai Biasa ---")
    pegawai_biasa.tampilkan_info()
    print(f"Total Gaji   : {format_rupiah(pegawai_biasa.hitung_gaji())}")
    print("-" * 35)
```

#Praktikum 6: Memeriksa Tipe Instance dengan isinstance()

```python
# --- Kelas Induk ---
class BentukGeometri:
    def __init__(self, nama_bentuk):
        self.nama = nama_bentuk

    def tampilkan_nama(self):
        print(f"Ini adalah bentuk: {self.nama}")

    def hitung_luas(self):
        print(f"Luas {self.nama} tidak dapat dihitung (generik).")

# --- Kelas Anak 1 ---
class PersegiPanjang(BentukGeometri):
    def __init__(self, panjang, lebar):
        # Memanggil konstruktor induk
        super().__init__("Persegi Panjang")
        self.panjang = panjang
        self.lebar = lebar

    def hitung_luas(self):
        luas = self.panjang * self.lebar
        print(f"Luas {self.nama}: {luas}")

# --- Kelas Anak 2 ---
class Segitiga(BentukGeometri):
    def __init__(self, alas, tinggi):
        # Memanggil konstruktor induk
        super().__init__("Segitiga")
        self.alas = alas
        self.tinggi = tinggi

    def hitung_luas(self):
        luas = 0.5 * self.alas * self.tinggi
        print(f"Luas {self.nama}: {luas}")

# --- Fungsi Helper untuk Mengecek Tipe Objek ---
def proses_bentuk(bentuk):
    print(f"\nMemproses objek: {type(bentuk)}")

    # Mengecek apakah objek adalah turunan dari BentukGeometri
    if isinstance(bentuk, BentukGeometri):
        bentuk.tampilkan_nama()
        print("Objek ini adalah instance dari BentukGeometri (atau turunannya).")

        # Pengecekan spesifik kelas anak
        if isinstance(bentuk, PersegiPanjang):
            print("-> Objek ini juga instance dari PersegiPanjang.")
        elif isinstance(bentuk, Segitiga):
            print("-> Objek ini juga instance dari Segitiga.")
    else:
        print("Objek ini BUKAN instance dari BentukGeometri.")

    # Mengecek jika objek adalah tipe data string
    if isinstance(bentuk, str):
        print("Objek ini adalah string.")

# --- Kode Utama ---
if __name__ == "__main__":
    # Inisialisasi berbagai objek
    pp = PersegiPanjang(4, 3)
    seg = Segitiga(5, 2)
    bg = BentukGeometri("Lingkaran")
    teks = "Ini string"

    # Masukkan ke dalam daftar untuk diproses secara otomatis
    daftar_objek = [pp, seg, bg, teks]

    for obj in daftar_objek:
        proses_bentuk(obj)
```

#Penugasan

```python
# --- Kelas Induk ---
class Mahasiswa:
    def __init__(self, nama, nim, ipk):
        self.nama = nama
        self.nim = nim
        self.ipk = ipk

    def tampilkan_info(self):
        print(f"Nama : {self.nama}")
        print(f"NIM  : {self.nim}")
        print(f"IPK  : {self.ipk}")

    def hitung_predikat(self):
        if self.ipk >= 3.5:
            predikat = "Dengan Pujian"
        elif self.ipk >= 3.0:
            predikat = "Sangat Memuaskan"
        elif self.ipk >= 2.5:
            predikat = "Memuaskan"
        else:
            predikat = "Cukup"
        print(f"Predikat: {predikat}")

# --- Kelas Anak 1 (Sarjana) ---
class MahasiswaSarjana(Mahasiswa):
    def __init__(self, nama, nim, ipk, semester, sks_lulus):
        super().__init__(nama, nim, ipk)
        self.semester = semester
        self.sks_lulus = sks_lulus

    def tampilkan_info(self):
        print("--- Info Mahasiswa Sarjana ---")
        super().tampilkan_info()
        print(f"Semester : {self.semester}")
        print(f"SKS Lulus: {self.sks_lulus}")
        self.hitung_predikat()

# --- Kelas Anak 2 (Magister) ---
class MahasiswaMagister(Mahasiswa):
    def __init__(self, nama, nim, ipk, judul_tesis, nama_pembimbing):
        super().__init__(nama, nim, ipk)
        self.judul_tesis = judul_tesis
        self.nama_pembimbing = nama_pembimbing

    def tampilkan_info(self):
        print("--- Info Mahasiswa Magister ---")
        super().tampilkan_info()
        print(f"Judul Tesis: {self.judul_tesis}")
        print(f"Pembimbing : {self.nama_pembimbing}")
        self.hitung_predikat()

# --- Kode Utama (Main) ---
if __name__ == "__main__":
    # Buat objek
    mhs_s1 = MahasiswaSarjana("Dewi Shinta", "240101", 3.85, 4, 80)
    mhs_s2 = MahasiswaMagister("Budi Pratama", "240901", 3.45, "Analisis IoT pada Smart City", "Dr. Ahmad")

    # Tampilkan data
    mhs_s1.tampilkan_info()
    print("-" * 30)
    mhs_s2.tampilkan_info()
```
